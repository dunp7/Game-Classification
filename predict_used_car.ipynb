```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('C:\\Users\\Administrator\\Desktop\\raw_data_crawled.csv')
df
```

```python
df = df.iloc[:,1:]
df
```

```python
df.price = df.price.str.replace(' tỉ ','')
df.price = df.price.str.extract('(\d+)')
df.nam_sx = df.nam_sx.str.extract('(\d+)')
df.origin = df.origin.str.replace('Xuất xứ ','')
df.type_car = df.type_car.str.replace('Kiểu dáng ','')
df.km_traveled = df.km_traveled.str.replace('.','')
df.km_traveled = df.km_traveled.str.extract('(\d+)')
df.gear = df.gear.replace('Hộp số ','')
df.condition = df.condition.str.replace('Tình trạng ','')
df.fuel = df.fuel.str.replace('Nhiên liệu ','')
df.name = df.name.str[7:-5]
df
```
